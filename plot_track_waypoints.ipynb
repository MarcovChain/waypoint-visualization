## Notebook to visualise waypoints for AWS DeepRacer tracks

### libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import pandas as pd
```

### load track name to visualise

```python
track_name = "Canada_Training" # reinvent2018

absolute_path = "./courses"
wp = np.load("%s/%s.npy" % (absolute_path, track_name))

centre = pd.DataFrame({'track': 'centre',
                        'x': wp[:,0],
                        'y': wp[:,1]
                    })

inner = pd.DataFrame({'track': 'inner',
                        'x': wp[:,2],
                        'y': wp[:,3]
                    })

outer = pd.DataFrame({'track': 'outer',
                        'x': wp[:,4],
                        'y': wp[:,5]
                    })         

tracks = outer.append(centre).append(inner)

fig = px.scatter(tracks, x="x", y="y", color="track",
                 hover_data=['track'])
fig.show()
```

### enumerate waypoints

```python
for i, point in enumerate(wp):
    waypoint = (point[2], point[3])
    print("Waypoint " + str(i) + ": " + str(waypoint))
```
